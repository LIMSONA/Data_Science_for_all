 * Reshaping: https://pandas.pydata.org/pandas-docs/stable/user_guide/reshaping.html#reshaping-stacking

### 1) 스택 stack

* 모든 데이터들을 인덱스 레벨로 변형합니다. 이를 압축(compresses)한다고 표현

```python
tuples = list(zip(*[['bar', 'bar', 'baz', 'baz', 'foo', 'foo', 'qux', 'qux'],
                    ['one', 'two', 'one', 'two', 'one', 'two', 'one', 'two']]))
tuples
```

* 멀티 인덱스(multiindex) : 말 그대로 index를 한 번에 여러 개 사용하는 방법

```python
index = pd.MultiIndex.from_tuples(tuples, names=['first', 'second'])
type(index)
```

```python
df = pd.DataFrame(np.random.randn(8, 2), index=index, columns=['A', 'B'])
df
```

```python
df2 = df[:4]
df2
```

```python
stacked = df2.stack()
stacked
```

```python
#타입 확인하기   ---> 시리즈!!

type(stacked)
```

* unstack
 - “stacked”된 DataFrame 이나 Series 를 원래 형태로 되돌릴 수 있습니다. 
 - 되돌리는(압축 해제) 것의 레벨을 정할 수 있습니다.

```python
# 레벨 -1 : 원래 값으로 되돌린다.

stacked.unstack()
```

```python
# 레벨 0은  : first 인덱스가 해제되어 >>> bar와 baz 컬럼이 생김
#인덱스 중 [0]이 first니까

stacked.unstack(0)
```

```python
# 레벨 1 : second 인덱스가 해제되어서 >>> one, two 칼럼이 생김
#인덱스중 [1]이 second니까

stacked.unstack(1)
```

## 9. 시계열 데이터 -->어렵...ㅠ

* 더보기 :  https://pandas.pydata.org/pandas-docs/stable/user_guide/timeseries.html#timeseries

* resample() :  데이터 샘플링하기

```python
import pandas as pd
# 1초 단위로 100개의 index 생성
# 100개 만들거니까 periods=100 / 초단위니까 freq = "S" 

rng = pd.date_range("2/22/2022", periods=100, freq="S")
rng
```

```python
ts = pd.Series(np.random.randint(0,500,len(rng)), index=rng)
ts
```

```python
# 5분 단위로 샘플링하여 sum()
ts.resample("5Min").sum()
```

```python
# 다양한 타임존으로 변경가능하다

rng = pd.date_range('2/22/2022 22:22', periods=5, freq='D')
rng
```

```python
ts = pd.Series(np.random.randn(len(rng)), rng)
ts
```

```python
# 표준시(UTC)로 변경
ts_utc = ts.tz_localize("UTC")
ts_utc
```

```python
# US 동부 시각으로 변경

ts_utc.tz_convert('US/Eastern')
```

```python
# 시간 간격(TimeSpan)도 쉽게 표현
rng = pd.date_range("1/1/2022", periods=5, freq="M")
ts = pd.Series(np.random.randn(len(rng)), index=rng)
ts
```

* 기간(period)과 시간(timestamp) 사이에 산술적인 기능들을 적용가능
 * 3시간씩 더하기 예제

```python
#1990~2000 년 분기별
prng = pd.period_range("1990Q1","2000Q4", freq="Q-NOV")
prng
```

```python
ts = pd.Series(np.random.randn(len(prng)), prng)
ts
```

```python
#3시간 더하기!
ts.index = (prng.asfreq('M', 'e') + 1).asfreq('H', 's') + 3
ts.index
```

## 10. 범주형 데이터

* 더보기
 * https://pandas.pydata.org/pandas-docs/stable/user_guide/categorical.html#categorical
 * https://pandas.pydata.org/pandas-docs/stable/reference/arrays.html#api-arrays-categorical

```python
# .astype("category")로 범주형 데이터 타입으로 변환

df = pd.DataFrame({"id":[1,2,3,4,5,6], "raw_grade":['a', 'b', 'b', 'a', 'a', 'c']})
df
```

```python
df["grade"]= df["raw_grade"].astype("category")
df.info()
```

```python
df["grade"].unique()
```

```python
# cat.categories 속성을 이용하여 카테고리명을 다시 만들 수 있음
# a-> very good / b-> good / c-> very bad

df["grade"].cat.categories = ["very good", "good", "very bad"]
df
```

```python
# 범주형 데이터를 정렬 : sort_values()

df.sort_values(by="grade")
```

```python
# 그룹화하여 빈도수를 출력하면, 비어있는 범주도 쉽게 확인가능

df.groupby("grade").size()
```
